```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 加载两张图片
img1 = mpimg.imread("figures/output1.png")
img2 = mpimg.imread("figures/output2.png")

# 创建一个图形和子图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # 1行2列的子图

# 在子图中显示图片
ax1.imshow(img1)
ax1.axis('off')  # 关闭坐标轴
ax1.set_title("Left Image")  # 添加标题（可选）

ax2.imshow(img2)
ax2.axis('off')  # 关闭坐标轴
ax2.set_title("Right Image")  # 添加标题（可选）

# 调整子图间距
plt.tight_layout()

# 显示结果
plt.show()

# 如果需要保存拼接后的图片
fig.savefig("output_combined.png", bbox_inches='tight')
```
